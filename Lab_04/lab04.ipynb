### 1. Класи суміжності та нормальні підгрупи

Розглянемо групу $S_4$ та її підгрупи:<br>
$\quad$ (a) $H_1=\langle (1,2,3,4)\rangle$;<br>
$\quad$ (b) $H_2=\langle (1,3), (1,2,3,4)\rangle$;<br>
$\quad$ (c) $H_3=K_4$.

Для кожної з цих підгруп $H$ виконайте наступні кроки:

1. Знайдіть порядок $H$ та її індекс $[S_4:H]$.
2. Перевірте, чи є підгрупа $H$ нормальною.  % H.is_normal(G)
3. Якщо $H$ є нормальною, то побудуйте факторгрупу $G/H$, її таблицю Келі та визначіть структуру.  % G.quotient(H), .structure_description()
4. Якщо $H$ не є нормальною, то знайдіть для неї ліві та праві класи суміжності.  % G.cosets(H) 



```python
G = SymmetricGroup(4)
sigma1 = G((1,2,3,4))
H1 = G.subgroup([sigma1])
print("order H1:", H1.order())
#print("[S4:H1] =", G.index(H1))
print("Index [S4:H1]:", G.order() // H1.order())
norm = H1.is_normal(G)
print("is H1 normal: ", norm)
if norm:
    Q = G.quotient(H1)
    print("G/H1:", Q)
    print(Q.cayley_table())
    print(Q.structure_description())
else:
    print("left cossets")
    for c in G.cosets(H1, side='left'):
        print(c)
    
    print("right cossets")
    for c in G.cosets(H1, side='right'):
        print(c)
        
sigma2 = G((1,3),(1,2,3,4))
H2 = G.subgroup([sigma2])
print("order H2:", H2.order())
print("Index [S4:H2]:", G.order() // H2.order())
norm = H2.is_normal(G)
print("is H2 normal: ", norm)
if norm:
    Q = G.quotient(H2)
    print("G/H2:", Q)
    print(Q.cayley_table())
    print(Q.structure_description())
else:
    print("left cossets")
    for c in G.cosets(H2, side='left'):
        print(c)
    
    print("right cossets")
    for c in G.cosets(H2, side='right'):
        print(c)

        
a = G("(1,2)(3,4)")
b = G("(1,3)(2,4)")

K4 = G.subgroup([a, b])
print("order K4:", K4.order())
print("Index [S4:K4]:", G.order() // K4.order())
norm = K4.is_normal(G)
print("is K4 normal: ", norm)
if norm:
    Q = G.quotient(K4)
    print("G/K4:", Q)
    print(Q.cayley_table())
    print(Q.structure_description())
else:
    print("left cossets")
    for c in G.cosets(K4, side='left'):
        print(c)
    
    print("right cossets")
    for c in G.cosets(K4, side='right'):
        print(c)
```

### 3. Проекції групи $SL_2(\mathbb{Z}_m)$ на $SL_2(\mathbb{Z}_n)$ 

Для значень $m=6,8,20,30,100$ задайте групу $SL_2(\mathbb{Z}_m)$.
Для кожного дільника $n$ числа $m$:

1. Задайте групу $SL_2(\mathbb{Z}_n)$.
2. Задайте гомоморфізм з групи $SL_2(\mathbb{Z}_m)$ в групу $SL_2(\mathbb{Z}_n)$, який редукує матриці mod n. % M.change\_ring\(Integers\(n\)\)
3. Перевірте, чи є гомоморфізм сюр'єктивним.
4. Знайдіть порядок ядра гомоморфізму та порівняйте з $|SL_2(\mathbb{Z}_m)| / |SL_2(\mathbb{Z}_n)|$.<br>



```python
def to_mat_over(M, R):
    # перетворити будь-який matrix-like об'єкт у матрицю над R
    return Matrix(R, [[R(x) for x in row] for row in matrix(M)])

m_values = [6, 8, 20, 30, 100]

for m in m_values:
    print("\n" + "="*40)
    print("m =", m)
    R1 = IntegerModRing(m)
    G1 = SL(2, R1)           # <= скінченна група SL(2, Z/mZ)
    gens = G1.gens()

    for n in divisors(m):
        if n == 1:           # опціонально пропускаємо мод 1
            print(" skip n=1")
            continue

        print("\n -> projection mod", n)
        R2 = IntegerModRing(n)
        G2 = SL(2, R2)

        # Побудова образів генераторів доменa у вигляді елементів G2
        images = []
        ok = True
        for g in gens:
            try:
                M = matrix(g)                # витягнути "звичайну" матрицю
                M2 = to_mat_over(M, R2)      # матриця над R2
                images.append(G2(M2))       # звести до елемента G2
            except Exception as e:
                print("  conversion failed for a generator:", e)
                ok = False
                break

        if not ok:
            print("  skip this (m,n) pair due to conversion problem")
            continue

        # Створюємо гомоморфізм G1 -> G2
        try:
            phi = G1.hom(images, G2)
            print("  Homomorphism constructed:", phi)
        except Exception as e:
            print("  Failed to construct homomorphism:", e)
            continue

        # Інформація (обережно: деякі операції важкі)
        try:
            print("   |domain| =", G1.order())
            print("   |codomain| =", G2.order())
        except Exception:
            print("   couldn't compute orders")

        # Перевірка сюр'єктивності (якщо можливо)
        try:
            Im_sub = G2.subgroup(images)       # підгрупа образів
            im_order = Im_sub.order()          # порядок збудованої підгрупи
            print("   image order (via subgroup):", im_order)
            print("   surjective?:", im_order == G2.order())
        except Exception as e_sub:
            print("   G2.subgroup failed:", e_sub)
            print("   Fallback: build image by closure of generators (may be slow)...")

            # --- Запасний ручний спосіб: closure (BFS) ---
            try:
                # images: список елементів G2 (елементи-класу групи)
                gens_img = images[:]  # копія
                # множина для зберігання унікальних елементів; використовуємо str() ключ для надійності
                seen = { g for g in gens_img }           # set елементів (еквівалентно)
                frontier = list(gens_img)

                # розширюємо підгрупу поки приходять нові елементи
                while frontier:
                    x = frontier.pop()
                    for g in gens_img:
                        for y in (x * g, g * x):   # праві і ліві добутки (щоб швидше замкнути)
                            if y not in seen:
                                seen.add(y)
                                frontier.append(y)

                im_order = len(seen)
                print("   image order (via closure):", im_order)
                print("   surjective?:", im_order == G2.order())
            except Exception as e_cl:
                print("   Fallback closure failed:", e_cl)

        # Ядро (може бути важко)
        try:
            K = phi.kernel()
            print("   kernel order:", K.order())
        except Exception as e:
            print("   Could not compute kernel:", e)
```

### 6*. Автоморфізми симетричних груп $S_n$

Для майже всіх симетричних груп $S_n$ усі автоморфізми є внутрішніми, однак існує одне особливе значення 
$n$, для якого з’являється зовнішній автоморфізм.

1. Для кожного $n=2,3,4,\ldots,8$:<br>

   (б) Знайдіть групу автоморфізмів $Aut(S_n)$ та внутрішніх автоморфізмів $Inn(S_n)$.<br>
   (в) Порівняйте їх порядки та визначте, для якого $n$ вони відрізняються.<br>
2. Для знайденого значення $n$:<br>
   (a) Побудуйте невнутрішній автоморфізм групи $S_n$.<br>
   (б) Покажіть, як він діє на твірних групи (наприклад, на сусідніх транспозиціях $(1\,2), (2\,3),\ldots, (n-1\,n)$.



```python
def check(n):
    G = SymmetricGroup(n)
    g = gap(G) 
    Aut = g.AutomorphismGroup()
    Inn = g.InnerAutomorphismGroup()
    print(f"n = {n}")
    print(" |S_n| =", g.Size())
    print(" |Aut| =", Aut.Size())
    print(" |Inn| =", Inn.Size())
    print(" ratio =", Aut.Size() / Inn.Size())
    #iso = gap.eval("IsIsomorphic(aut, inn);")
    #iso = str(gap("IsIsomorphic(Aut, Inn);")).strip()
    #print(" Aut == Inn ?", int(str(gap("Size(Aut);"))) == int(str(gap("Size(Inn);"))))
    
    sizeAut = int(str(Aut.Size()))
    sizeInn = int(str(Inn.Size()))
    
    print(" Aut == Inn ?", sizeAut == sizeInn)
    #print(" Aut ≅ Inn ?", iso)
    print("")

for n in range(2,8):
    check(n)

    
```

```python
from sage.interfaces.gap import gap
from sage.groups.perm_gps.permgroup_named import SymmetricGroup

# Створюємо S6 у Sage
G = SymmetricGroup(6)

# Передаємо об'єкт Sage у GAP
G_gap = gap(G)

# Обчислюємо автоморфізми та внутрішні автоморфізми через GAP
AutG = G_gap.AutomorphismGroup()
InnG = G_gap.InnerAutomorphismGroup()

# Порівнюємо порядки
print("|S_6|       =", G.order())
print("|Aut(S_6)|  =", int(AutG.Size()))
print("|Inn(S_6)|  =", int(InnG.Size()))
print("Aut == Inn? ", int(AutG.Size()) == int(InnG.Size()))

# RightTransversal AutG/InnG для пошуку зовнішнього автоморфізму
transv = AutG.RightTransversal(InnG)
print("Outer automorphism exists?", len(transv) > 1)
print("Transversal elements (GAP objects):")
for t in transv:
    print(t)
```
