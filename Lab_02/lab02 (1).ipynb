### 6*. Група підстановок, породжена випадковими підстановками

1. Створіть групи $S_n$ та $A_n$ для великого числа $n$.
2. Проведіть наступний експеримент багато разів (в циклі):<br>
       2.1. Згенеруйте дві випадкові підстановки в $S_n$.<br>
       2.2. Перевірте, парні вони чи ні.<br>
       2.3. Перевірте, чи породжують вони всю групу $S_n$ або її підгрупу $A_n$. 
4. Сформулюйте гіпотезу про те, яка ймовірність породження симетричної групи двома підстановками та що відбувається при $n\rightarrow\infty$.

```python
n = 10
G = SymmetricGroup(n)
A = AlternatingGroup(n)

num_trials = 20

count_generate_Sn = 0
count_generate_An = 0
count_other = 0

def is_even_perm_simple(p, n):
    perm = [p(i) for i in range(1, n+1)]   # очікується, що p(i) дає образ i у {1..n}
    inv = 0
    for i in range(n):
        for j in range(i+1, n):
            if perm[i] > perm[j]:
                inv += 1
    return (inv % 2) == 0


for i in range(num_trials):
    # дві випадкові підстановки
    g1 = G.random_element()
    g2 = G.random_element()
    
    # парні вони чи ні
    is_g1_even = is_even_perm_simple(g1, n)
    is_g2_even = is_even_perm_simple(g2, n)
    
    # яку групу породжують
    H = G.subgroup([g1, g2])
    
    if H == G:
        count_generate_Sn += 1
    elif H == A:
        count_generate_An += 1
    else:
        count_other += 1
    
    print(f"Експеримент {i+1}:")
    print(f"  g1 парна? {is_g1_even}, g2 парна? {is_g2_even}")
    print(f"  Порядок породженої підгрупи: {H.order()}")
    print("")

# Підсумок 
print("Результати")
print(f"Загальна кількість експериментів: {num_trials}")
print(f"Породили всю S_n: {count_generate_Sn} разів")
print(f"Породили A_n:     {count_generate_An} разів")
print(f"Інше:             {count_other} разів")

p_Sn = float(count_generate_Sn / num_trials)
p_An = float(count_generate_An / num_trials)

print(f"Ймовірність породження S_n ≈ {p_Sn:.3f}")
print(f"Ймовірність породження A_n ≈ {p_An:.3f}")
```

### 7*. Чи є граф Петерсена графом Келі?

1. Побудуйте граф Петерсена.
2. Перевірте, чи є граф Петерсена графом Келі деякої групи.

```python
#Побудова
P = graphs.PetersenGraph()
print("Petersen graph: n =", P.num_verts(), "m =", P.num_edges())

# явний список вершин у вигляді 0..9
verts = list(P.vertices())
print("Вершини:", verts)

print("Чи є граф вершинно-транзитивним? ->", P.is_vertex_transitive())
print("Чи є граф регулярним? ->", P.is_regular())
print("Ступінь кожної вершини:", P.degree(0))

# Автоморфізми
Aut = P.automorphism_group()
print("Група автоморфізмів Aut(P):", Aut)
print("Порядок групи автоморфізмів |Aut(P)| =", Aut.order())

# Шукаємо регулярну підгрупу розміру 10
target_order = P.num_verts()   # 10
found_regular = False
regular_subgroups = []

# Отримаємо підгрупи автоморфізмів
subs = Aut.subgroups()

print("Знайдено підгруп підгруп:", len(subs))
for H in subs:
    if H.order() != target_order:
        continue
    # кожен нетривіальний елемент H не фіксує жодної вершини?
    H_elems = H.list()
    regular = True
    for g in H_elems:
        if g == Aut.identity():   # ідентичність
            continue
        # існує вершина v така, що g(v) == v?
        fixed_found = False
        for v in verts:
            # Примітка: елементи Aut — PermutationGroupElement, їх можна застосувати до індексів
            # якщо вершини не 0..n-1, можливо потрібно зіставити. У стандартному PetersenGraph вершини - 0..9
            image = g(v)
            if image == v:
                fixed_found = True
                break
        if fixed_found:
            regular = False
            break
    if regular:
        found_regular = True
        regular_subgroups.append(H)

# Вивід результату
if found_regular:
    print("Знайдена регулярна підгрупа порядку", target_order)
    for H in regular_subgroups:
        print("  - підгрупа H, порядок:", H.order(), "генератори:", H.gens())
else:
    print("Не знайдено регулярної підгрупи порядку", target_order)
    print("Отже, Petersen graph НЕ є графом Келі.")
```
